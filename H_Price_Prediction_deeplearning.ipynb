```python
import pandas as pd
```

```python
data=pd.read_csv('/Housing.csv')
data.head()
data.shape
```

```python
feature = ['area','bedrooms','bathrooms','mainroad','guestroom','hotwaterheating','airconditioning','parking','furnishingstatus']
x=data[feature]
y=data.price
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
le=LabelEncoder()
x=x.apply(le.fit_transform)
x
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(5,input_shape=(x_train.shape[1],))
])
```

```python
# prompt: how  to ADd layers

model.add(tf.keras.layers.Dense(10, activation='relu'))
model.add(tf.keras.layers.Dense(1)) # Output layer
```

```python
#compile the model
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse'])
```

```python
#training of model
hist=model.fit(x_train,y_train,epochs=150,validation_split=0.2)
```

```python
#evaluation of model
model.evaluate(x_test,y_test)
```

```python
model.predict(x_test)
```
